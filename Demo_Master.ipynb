Accurate earthquake predictions could save lives and property, but so far no method can predict the location, time, and magnitude precisely enough to be useful. One of the best predictive models is ETAS: Epidemic Type AfterShocks. ETAS exploits the empirical clustering of earthquakes to predict more accurately than many simpler models, but it has no physical basis. Professors Brad Luen and Philip B. Stark previously found that Magnitude Dependant Alarms (MDA) did essentially as well as the Epidemic Type Aftershock Sequence (ETAS) model but is much simpler. 

As the collaborative project for Statistics 157: Collaborative and Reproducible Data Science, we worked together to find a method to predict earthquakes that takes advantage of clustering to performs better than ETAS and MDA but was still simple. Performance is measured by the area under the error diagram, which plots the fraction of events that are not predicted versus the fraction of the observation period during which “alarms” are active.

# Model 1: MDA model generation and resulting plot

We wanted to modify the MDA model by reducing the window size slightly for earthquakes with bigger magnitudes. Because the μ^Μ portion in Luen’s model already scaled the window size to increase exponentially, we don’t want the window to be too big for large magnitude earthquakes. So we might want to scale down the window size slightly for big earthquakes slightly so that we still capture aftershocks but reduce the number of false positives or times when the alarm is on but no earthquakes occur. In this way we might be able to reduce the tau for the same v given in Luen’s model.

# Model 2: Empirical Modeling with Percentiles of Inter-Arrival Times 

```python
#Quantiles Demo
quantile_code = quantile_util.generate_quantile_plots()
#add plotting here
```

```python
#image 1
Image(filename=os.path.join(HOME_DIR, 'tristan_quakers/Quantile-Method/90Percentile.jpeg'), width=550, height=550)
```

The graph above shows the 90th percentile of the waiting time until an event of magnitude 3 or larger after an event of a given magnitude, for Southern California Earthquake Center data for 1983-2012. Our original idea was to use these empirical percentiles to decide the length of an alarm. However, considering the possibility that a longer window after big earthquakes may capture more events, we decided to use an increasing function with constant slope of percentiles (from 70th percentile to 99th percentile) to determine the length of the alarms with respect to magnitude bins.

```python
#Quantile image 2 (result)
Image(filename=os.path.join(HOME_DIR, 'tristan_quakers/Quantile-Method/ErrorECDF.jpeg'), width=550, height=550)
```

- We found that the area under the curve is 0.2806, larger than the area under the MDA (0.2695) and ETAS error diagrams. 
- If we had more time, we might be able to modify this model better with an exponentially increasing function for the percentile cutoff points for each bin of magnitude.
